```python
from ml4floods.data import utils

aois = utils.read_geojson_from_gcp("/home/gonzalo/Downloads/map_north_island_NZ.geojson")
aois.explore()
```

```python
from datetime import datetime, timedelta, timezone

date_event = datetime.strptime("2023-02-13","%Y-%m-%d").replace(tzinfo=timezone.utc)

date_start_search = datetime.strptime("2023-01-20","%Y-%m-%d").replace(tzinfo=timezone.utc)
date_end_search = date_event + timedelta(days=20)
```

```python
from georeader.readers import query_utils
from georeader.readers import ee_query

flood_images_gee, flood_collection = ee_query.query(
    area=aois.unary_union, 
    date_start=date_start_search, 
    date_end=date_end_search,                                                   
    producttype="both", 
    return_collection=True, 
    add_s2cloudless=True)

flood_images_s1_gee, flood_collection_s1 = ee_query.query_s1(
    area=aois.unary_union, 
    date_start=date_start_search, 
    date_end=date_end_search,      
    return_collection=True)

num_images = flood_images_s1_gee.shape[0]
print(f"[INFO] Found {num_images} flooding images on S1 archive.")

num_images = flood_images_gee.shape[0]
print(f"[INFO] Found {num_images} flooding images on S2/L8/L9 archive.")
```

```python
import pandas as pd
flood_images_gee["orbitProperties_pass"] = "DESCENDING"
flood_images_gee_all = pd.concat([flood_images_gee,flood_images_s1_gee]).sort_values("localdatetime")
flood_images_gee_all
```

```python
# Format the date, columns to show and colours
import folium

flood_images_gee_all["localdatetime_str"] = flood_images_gee_all["localdatetime"].dt.strftime("%Y-%m-%d %H:%M:%S")

showcolumns = ["geometry","overlappercentage","cloudcoverpercentage", "localdatetime_str","solarday","satellite"]
showcolumns_s1 = ["geometry","overlappercentage","orbitProperties_pass","localdatetime_str","solarday","satellite"]
colors = ["#ff7777", "#fffa69", "#8fff84", "#52adf1", "#ff6ac2","#1b6d52", "#fce5cd","#705334"]

# Plot the AoI outline
m = aois.explore(style_kwds={"fillOpacity": 0.1}, color="black", name="AoI Outline")
    
# Add the flooding data
for i, ((day,satellite,orbit), images_day) in enumerate(flood_images_gee_all.groupby(["solarday","satellite","orbitProperties_pass"])):
    if satellite.startswith("S1"):
        m = images_day[showcolumns_s1].explore(
            m=m, 
            name=f"{satellite} ({orbit}): {day}", 
            color=colors[i % len(colors)], 
            show=False)
    else:
        m = images_day[showcolumns].explore(
            m=m, 
            name=f"{satellite}: {day}", 
            color=colors[i % len(colors)], 
            show=False)
    
# Add the layer control and show
folium.LayerControl(collapsed=False).add_to(m)
m
```
